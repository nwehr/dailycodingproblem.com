## #23 [Easy]

This problem was asked by Google.

You are given an `M` by `N` matrix consisting of booleans that represents a board. Each True boolean represents a wall. Each False boolean represents a tile you can walk on.

Given this matrix, a start coordinate, and an end coordinate, return the minimum number of steps required to reach the end coordinate from the start. If there is no possible path, then return null. You can move up, left, down, and right. You cannot move through walls. You cannot wrap around the edges of the board.

For example, given the following board:

```
[[f, f, f, f],
[t, t, f, t],
[f, f, f, f],
[f, f, f, f]]
```

and start = (3, 0) (bottom left) and end = (0, 0) (top left), the minimum number of steps required to reach the end is 7, since we would need to go through (1, 2) because there is a wall everywhere else on the second row.

````python
UP    = 0
DOWN  = 1
LEFT  = 2
RIGHT = 3

def up(position: (int, int)) -> (int, int):
    x, y = position
    return (x, y - 1)

def down(position: (int, int)) -> (int, int):
    x, y = position
    return (x, y + 1)

def left(position: (int, int)) -> (int, int):
    x, y = position
    return (x - 1, y)

def right(position: (int, int)) -> (int, int):
    x, y = position
    return (x + 1, y)

class Board:
    def __init__(self, matrix: [[bool]]):
        self.matrix = matrix
        
    def can_move(self, position: (int, int)) -> bool:
        x, y = position
        
        if y < 0 or x < 0 or y == len(self.matrix) or x == len(self.matrix[y]) or self.matrix[y][x]:
            return False
        
        return True
        
    def min_distance(self, start, end: (int, int)) -> int:
        distances = set()
        
        def in_path(position: (int, int), path: [(int, int)]) -> bool:
            for p in path:
                if p == position:
                    return True

            return False
        
        def distance(position, end: (int, int), path: [(int, int)]):
            if position == end:
                distances.add(len(path))
                return

            x, y = position

            if self.can_move(up(position)) and not in_path(up(position), path):
                distance(up(position), end, [*path, position])

            if self.can_move(down(position)) and not in_path(down(position), path):
                distance(down(position), end, [*path, position])

            if self.can_move(left(position)) and not in_path(left(position), path):
                distance(left(position), end, [*path, position])

            if self.can_move(right(position)) and not in_path(right(position), path):
                distance(right(position), end, [*path, position])
        
        distance(start, end, [])
        
        return min(distances)

matrix: [[bool]] = [
    [False, False, False, False],
    [True, True, False, True],
    [False, False, False, False],
    [False, False, False, False],
]

board = Board(matrix)
print(board.min_distance((3, 0), (0, 0)))
print(board.min_distance((0, 3), (0, 0)))
````
