## #26 [Medium]

This problem was asked by Google.

Given a singly linked list and an integer k, remove the kth last element from the list. k is guaranteed to be smaller than the length of the list.

The list is very long, so making more than one pass is prohibitively expensive.

Do this in constant space and in one pass.

```python
class Item:
    pass

class Item:
    def __init__(self, value: int, next: Item = None):
        self.value = value
        self.next = next
        
        
class LinkedList:
    def __init__(self):
        self.head: Item = None
        self.tail: Item = None
            
    def add(self, value: int):
        if self.head == None:
            self.head = Item(value)
            self.tail = self.head
            return
            
        self.tail.next = Item(value)
        self.tail = self.tail.next
        
    def remove_kth_last(self, k: int) -> bool:
        preceeding_item = self.head
        current_item = self.head
        
        distance = 0
        
        while current_item.next:
            if distance > k:
                preceeding_item = preceeding_item.next
            
            current_item = current_item.next
            
            distance += 1
            
        if distance < k:
            return False
        
        if distance == k:
            self.head = self.head.next
        else:
            preceeding_item.next = preceeding_item.next.next
            
        return True
            
    def to_list(self) -> [int]:
        result = []
        
        current_item = self.head
        
        while current_item:
            result.append(current_item.value)
            current_item = current_item.next
        
        return result
        
my_list = LinkedList()
my_list.add(1)
my_list.add(2)
my_list.add(3)
my_list.add(4)
my_list.add(5)

assert my_list.to_list() == [1, 2, 3, 4, 5]
assert my_list.remove_kth_last(0)
assert my_list.to_list() == [1, 2, 3, 4]
assert my_list.remove_kth_last(1)
assert my_list.to_list() == [1, 2, 4]
assert my_list.remove_kth_last(100) == False
assert my_list.to_list() == [1, 2, 4]
```
